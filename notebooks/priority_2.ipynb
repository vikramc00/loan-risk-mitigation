```python
def getp2(cid, df, label):
    p2 = pd.DataFrame()

    p2['date'] = df['date']
    p2['current_ratio'] = df['current_assets'] / df['current_liabilities']
    p2['debt_equity_ratio'] = (df['current_liabilities'] + df['non_current_liabilities']) / (df['equity'])
    p2['acid_ratio'] = df['qr_top'] / df['current_liabilities']
    p2['net_profit_margin'] = (df['net_income'] / df['sales']) * 100
    p2['return_on_equity'] = df['net_income'] / df['equity'] 
    p2['debt_asset_ratio'] = (df['current_liabilities'] + df['non_current_liabilities']) / (df['current_assets'] + df['non_current_assets'])
    p2['gross_profit_margin'] = (df['sales'] - df['expenses']) / (df['sales'])
    p2.replace([np.inf, -np.inf], np.nan, inplace=True)
    p2 = p2.fillna(0)
    
    p2.to_csv("../Features/" + str(cid) + "/priority2_" + label + ".csv")
```
